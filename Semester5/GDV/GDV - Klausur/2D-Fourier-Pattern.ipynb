```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def ft2(img):
    d = np.fft.fftshift(np.fft.fft2(img))
    
    a = d.real
    b = d.imag
    
    amplitude = np.sqrt(a**2 + b**2)
    phase = np.arctan2(b, a)
    
    max_val = amplitude.max()
    amplitude /= max_val
    
    d_img = np.log(amplitude) * 0.1 + 1.0
    d_img[d_img < 0] = 0
    
    return d_img
    
```

```python
square = cv.imread('square.png', 0)
big_square = cv.imread('big_square.png', 0)
rotated_square = cv.imread('rotated_square.png', 0)
horizontal_rect = cv.imread('horizontal_rectangle.png', 0)
vertical_rect = cv.imread('vertical_rectangle.png', 0)
mult_sqares = cv.imread('mult_squares.png', 0)
square_pattern = cv.imread('square_pattern.png', 0)
hexagon = cv.imread('hexagon.png', 0)
bezier = cv.imread('bezier.png', 0)
big_dot = cv.imread('big_dot.png', 0)
dot = cv.imread('dot.png', 0)
dots = cv.imread('dots.png', 0)
empty_triangle = cv.imread('empty_triangle.png', 0)
triangle = cv.imread('triangle.png', 0)
horizontal_ellipse = cv.imread('horizontal_ellipse.png', 0)
vertical_ellipse = cv.imread('vertical_ellipse.png', 0)
vertical_slit = cv.imread('small_vertical_slit.png', 0)
big_vertical_slit = cv.imread('big_vertical_slit.png', 0)
```

```python
plt.figure()

figure, cell = plt.subplots(4, 2, figsize=(15, 15)) # (reihen, spalten, gesamte Plotgre) -> returns figure and axes (cells)
cell[0][0].imshow(square, cmap='gray')
cell[0][0].set_title('spatial domain')
cell[0][0].get_xaxis().set_visible(False)
cell[0][0].get_yaxis().set_visible(False)

cell[0][1].imshow(ft2(square), cmap='gray')
cell[0][1].set_title('frequency domain')
cell[0][1].axis('off')

cell[1][0].imshow(big_square, cmap='gray')
cell[1][0].get_xaxis().set_visible(False)
cell[1][0].get_yaxis().set_visible(False)

cell[1][1].imshow(ft2(big_square), cmap='gray')
cell[1][1].axis('off')

cell[2][0].imshow(dot, cmap='gray')
cell[2][0].get_xaxis().set_visible(False)
cell[2][0].get_yaxis().set_visible(False)


cell[2][1].imshow(ft2(dot), cmap='gray')
cell[2][1].axis('off')


cell[3][0].imshow(big_dot, cmap='gray')
cell[3][0].get_xaxis().set_visible(False)
cell[3][0].get_yaxis().set_visible(False)

cell[3][1].imshow(ft2(big_dot), cmap='gray')
cell[3][1].axis('off')



plt.show()
```
